```python
import tensorflow as tf
print(tf.__version__)
```

```python
a=tf.constant(10)
b=tf.constant(20)
print(a+b)

```

```python
v=tf.constant([1,2,3,4])
```

```python
t3 = tf.constant([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print(t3)
```

```python
a1=tf.constant([1,2,3,4])
print(a1.dtype,a1.shape,tf.rank(a1))
```

```python
a=tf.constant([10,20,30])
b=tf.constant([50,60,70])
c=tf.add(a,b)
print(c)

d=tf.subtract(a,b)
print(d)

e=tf.multiply(a,b)
print(e)

f=tf.divide(a,b)
print(f)

g=tf.pow(a,2)
print(g)
```
